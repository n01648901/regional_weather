```python
import requests

# List of cities to fetch weather data for
cities = ['Toronto', 'Montreal', 'Calgary', 'Vancouver']


# API key for weather API
api_key = '7d914127554a4db9be0223510240606'



# Base URL for weather API
base_url = 'https://api.weatherapi.com/v1/forecast.json'
```

```python
# Function to fetch weather data for a given city
def fetch_weather_data(city):
    # Parameters for the API request
    params = {
        'key': api_key,
        'q': city,
        'days': 7,
        'aqi': 'no',
        'alterts': 'no'
    }

    # Send GET request to the API
    response = requests.get(base_url, params=params)

    # Return the JSON response
    return response.json()


# Fetch weather data for all cities in the list
results = [fetch_weather_data(city) for city in cities]
```

```python
import pandas as pd
from pyspark.sql import SparkSession

# Create or get a Spark session
spark = SparkSession.builder.appName("WeatherData").getOrCreate()

# Convert the list of JSON responses to a Pandas DataFrame
df = pd.json_normalize(results)

# Display the DataFrame
df
```

```python
# Convert the Pandas DataFrame to a Spark DataFrame
spark_df = spark.createDataFrame(df)

# Display the Spark DataFrame
display(spark_df)
```
